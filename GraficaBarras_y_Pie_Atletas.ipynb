```python
import pandas as pd
import matplotlib.pyplot as plt

# Paso 2: Leer el archivo CSV y filtrar los datos para incluir solo a los atletas con medallas
datos_atletas = pd.read_csv('athlete_events2.csv')
datos_atletas_con_medalla = datos_atletas.dropna(subset=['Medal'])

# Paso 3: Agrupar los datos por país ('Team') y contar el número de medallas ganadas por cada país
medallas_por_pais = datos_atletas_con_medalla['Team'].value_counts()

# Seleccionar los 10 países con más medallas
top_10_paises = medallas_por_pais.head(10)

# Crear una gráfica de barras con los datos de los 10 países con más medallas
top_10_paises.plot(kind='bar', color='skyblue')

# Agregar etiquetas a los ejes y un título para la gráfica
plt.xlabel('País')
plt.ylabel('Número de Medallas')
plt.title('Top 10 Países con más Medallas')

# Mostrar la gráfica
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV y filtrar los datos para incluir solo a los atletas con medallas
datos_atletas = pd.read_csv('athlete_events2.csv')
datos_atletas_con_medalla = datos_atletas.dropna(subset=['Medal'])

# Obtener el año más reciente presente en los datos
ultimo_anio = datos_atletas['Year'].max()

# Filtrar los datos para incluir solo el año más reciente
datos_ultimo_anio = datos_atletas[datos_atletas['Year'] == ultimo_anio]

# Agrupar los datos por país y contar el número de medallas ganadas por cada país en el año más reciente
medallas_ultimo_año = datos_ultimo_anio['Team'].value_counts()

# Crear una gráfica de pastel con los datos de las medallas ganadas por cada país en el año más reciente
medallas_ultimo_año.plot(kind='pie')

# Agregar título para la gráfica
plt.title(f'Número de Medallas Ganadas por País en el Año {ultimo_anio}')

# Mostrar la gráfica
plt.show()
```
